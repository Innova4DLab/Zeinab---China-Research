```python
#Import Pandas package which help us to work with a structured dataset easily
import pandas as pd
import numpy as np
#For displaying a Pandas table in the jupyter notebook
from IPython.display import display, Markdown
#Read Microsoft Excel file
excel = pd.ExcelFile("2015_weibodata.xlsx")
#Load the dataset from the file as a Pandas table (called also dataframe)
data= excel.parse(0) #Obtain first sheet from the Excel file
#Show the first five rows of the table
data.head(5)
```

```python
# Check if there are null columns
data.isnull().any()
```

```python
# Eliminate columns that doesn't have information
cleanData = data[data.columns[0:17]]

cleanData.info()
cleanData.describe().transpose()
```

```python
cleanData.head(5)
```

```python
# Convert column names to alphanumeric

def filterColName(x):   
    import re
    return ''.join([n for n in x if (re.match("[a-zA-Z_-]", n))])

for column in cleanData.columns:
    cleanData = cleanData.rename(columns={column:filterColName(column)})
```

```python
#Transform province and city columns from string to integer 

def strToInt(x):
    x = str(x)
    return int(''.join([n for n in x if n.isdigit()]))


cleanData['province'] = cleanData['province'].apply(strToInt)
cleanData['city']=cleanData['city'].apply(strToInt)
cleanData.head(5)
```

```python
# http://open.weibo.com/wiki/Location/citycode/en

# Province
uniqueProvinces = np.unique(cleanData['province'].values)
print("{} unique provinces".format(len(uniqueProvinces)))

print(uniqueProvinces)

# Cities
uniqueCities = np.unique(cleanData['city'].values)
print("{} unique cities".format(len(uniqueCities)))

print(uniqueCities)
```

# From which parts of China people are more likely to socialize.




```python
import matplotlib.pyplot as plt

def plotPie(labels, sizes):
    %matplotlib inline
    #Labels associated to each gender
    #Number of rows related to each gender
    explode = (0, 0.1)  # only "explode" the 2nd slice
    fig1, ax1 = plt.subplots()
    #Plot the figure
    ax1.pie(sizes, 
            #explode=explode,
            labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    #Show the figure
    plt.show()
    
# https://pythonspot.com/matplotlib-bar-chart/
def plotBar(title, label_x, label_y, labels, data, fig_width=15 ):
    fig_height = (len(data) * .25)
    %matplotlib inline
    plt.figure(figsize=(fig_width,fig_height))
    y_pos = np.arange(len(labels))
    plt.barh(y_pos, data, align='center', alpha=0.5)
    plt.yticks(y_pos, labels)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.title(title)
    plt.show()
```

```python

# Weibos per city
weibosPerCity = cleanData.groupby(['city'])['city'].agg('count') #value_counts()
plotBar("Weibos per city", "# Weibos", "City code", [str(cityCode) for cityCode in weibosPerCity.index.tolist()],
       [weibosPerCity.loc[cityCode] for cityCode in weibosPerCity.index.tolist()])

# Weibos per province
weibosPerProvince = cleanData.groupby(['province'])['province'].agg('count') #value_counts()
plotBar("Weibos per province", "# Weibos", "Province code", [str(pCode) for pCode in weibosPerProvince.index.tolist()],
       [weibosPerProvince.loc[pCode] for pCode in weibosPerProvince.index.tolist()])

#Group Weibos by city and province
cityProvince = cleanData.groupby(['city', 'province'])['province'].agg('count')#.value_counts()
plotBar("Weibos per city/province", "# Weibos", "Province code", [str(pCode) for pCode in cityProvince.index.tolist()],
       [cityProvince.loc[pCode] for pCode in cityProvince.index.tolist()])
```

```python
genderTotal = cleanData.groupby(['gender'])['gender'].agg('count')

plotPie([gender.upper() for gender in genderTotal.index.tolist()],
       [genderTotal.loc[gender] for gender in genderTotal.index.tolist()])
```

# Create a graph where users from Sina Weibo represent vertices and the edges are created based on the proximity (a threshold) of users using lat/lon

```python
# For performance purpouses we get only a random subset from the original dataset and save into geoData variable

#Obtain a random row subset of size n from the original table #Obtain 
geoData = cleanData.sample(n=2000)

#Reset table subset indices (start from 0 to n-1) otherwise the table keeps original ones
geoData = geoData.reset_index(drop=True)
```

```python
# Distinct users
nodes = list(geoData.groupby(['user_id']).groups.keys())
print("Number of unique nodes:", len(nodes))
```

```python
import scipy.spatial as spatial

def get_edges(data, umbral = .001):
    # Get all points
    points = data[['lat', 'lon']]
    point_tree = spatial.cKDTree(points.values)
    
    locations = data.groupby(['user_id'])[['lat', 'lon']]
    for user_id in list(data.groupby(['user_id']).groups.keys()):
        user_locations = locations.get_group(user_id).values.tolist() # User locations (weibo gps coordinates)
        neighbour_indexes = point_tree.query_ball_point( user_locations, umbral)# Get neighbour locations
        neighbour_indexes = np.hstack(neighbour_indexes) # Squash n-dimensional (one array per weibo user_location) array
        # Frequency by neighbour location (can be the edge weight ...)
        neighbourWeight = {}
        for neighbour_index in neighbour_indexes:
            neighbour = data.iloc[neighbour_index][7]
            if user_id != neighbour:
                if str(neighbour) not in neighbourWeight:
                    neighbourWeight[str(neighbour)] = 0
                neighbourWeight[str(neighbour)] += 1

        for neighbour_user_id, frequency in neighbourWeight.items():
            #print(user_id, ",", neighbour_user_id, "," ,frequency)
            yield (user_id, int(neighbour_user_id) ,frequency)


# Generate edges based on the proximity of user weibos
for edge in [next(get_edges(geoData)) for _ in range(5)]:
    print(edge)
    
```

```python
# Export nodes and edges to GML/GEXF format in order to load into Gephi for best visualization (https://networkx.github.io/documentation/networkx-1.10/reference/generated/networkx.readwrite.gml.write_gml.html)
import networkx as nx
G=nx.Graph()

for node_id in nodes:
    G.add_node(node_id)

for edge in get_edges(geoData):
    if(edge[0] in nodes and edge[1] in nodes):
        G.add_edge(edge[0], edge[1], weight=edge[2])

#nx.draw(G)
nx.write_gml(G, "test.gml")
#nx.write_gexf(G, "test.gexf")
```

```python
# Interactive network visualization
from pyvis.network import Network

got_net = Network(height="1200px", 
                  width="100%", 
                  font_color="black", 
                  notebook=True)

for node_id in nodes:
    got_net.add_node(node_id, title=str(node_id))

for edge in get_edges(geoData):
    if(edge[0] in nodes and edge[1] in nodes):
        got_net.add_edge(edge[0], edge[1], weight=edge[2])

neighbor_map = got_net.get_adj_list()

# add neighbor data to node hover data
for node in got_net.nodes:
    node["title"] = " Neighbors:<br>" + "<br>".join( [str(i) for i in neighbor_map[node["id"]]] )
    node["value"] = len(neighbor_map[node["id"]])


got_net.show("graph.html") 
```
